### verify.py 불러오고, 안의 함수 사용
- load_models() : VideoMAE(행동), Whisper(음성) 불러옴
    - 리턴 : video_model, whisper_model, device
- action_verification() : VideoMAE를 이용해 행동 검증
    - 임계값(threshold) : 0.55로 설정
    - input : video_path, action_prompt, VideoMAE_model, device
    - return : True/False, 유사도 값
- speech_verification() : Whisper를 이용해 발화 검증
    - 임계값(tolerance) : 0.889로 설정
    - input : audio_path, speech_prompt, Whisper_model
    - return : True/False, 유사도 값
